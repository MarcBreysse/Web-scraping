```python
url='https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000&punto=indice'
```

```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import requests
from time import sleep
```

```python
def get_table(round, url=url):
    round_url = f'{url}/{round}'
    page = requests.get(round_url)
    soup = BeautifulSoup(page.text, 'html.parser')

    rows = []
    for child in soup.find_all('table')[4].children:
        row = []
        for td in child:
            try:
                row.append(td.text.replace('\n', ''))
            except:
                continue
        if len(row) > 0:
            rows.append(row)

    df = pd.DataFrame(rows[1:], columns=rows[0])

    return df

for round in range(1, 2):
    table = get_table(round)
    table.to_csv('MadridStockexchange_StocksQuotes.csv', index=False)
    sleep(np.random.randint(1, 10))

```

<span style=color:green>Check the latest price of the INDRA share.</span>

```python
# import libraries
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime
```

```python
# path to the web page
url = 'https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000&punto=indice'
```

```python
# request to process HTML web page
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Specify the table ID
tabla_result = soup.find('table', attrs={'id': 'ctl00_Contenido_tblAcciones'})
print(tabla_result.prettify())
```

```python
name=""
price=""
nroFila=0
for fila in tabla_result.find_all("tr"):
    if nroFila==23:
        nroCelda=0
        for celda in fila.find_all('td'):
            if nroCelda==0:
                name=celda.text
                print("Indice:", name)
            if nroCelda==1:
                price=celda.text
                print("Valor:", price)
            nroCelda=nroCelda+1
    nroFila=nroFila+1
```

```python
# Save information in CSV file
with open('bolsa_INDRAshare.csv', 'a') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow([name, price, datetime.now()])
```
